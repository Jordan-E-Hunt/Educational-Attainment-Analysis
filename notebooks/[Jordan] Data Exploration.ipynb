```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

#PATH CONFIGURATION
if 'google.colab' in str(get_ipython()):
    # If in Colab, we mount drive and point to the drive folder
    from google.colab import drive
    drive.mount('/content/drive')
    DATA_PATH = "/content/drive/MyDrive/Side Projects/Resume Projects/Berkeley College Project/[Black Alone] Educational Attainment.xlsx"
else:
    # If someone clones this from GitHub to their local machine:
    # It looks for the file in a relative 'data' folder
    DATA_PATH = 'data/[Black Alone] Educational Attainment.xlsx'

black_ed_data = pd.read_excel(DATA_PATH)
```

```python
##filename = "/content/drive/MyDrive/[Tuskegee] Data Deductors/College Preparatory Coursework.xls"
##CPC_data = pd.read_excel(filename)
```

```python
black_ed_data.replace({'-':np.nan},inplace=True)
black_ed_data = black_ed_data.replace(r'\t', '', regex=True)
rowlist = ["", "Total", "None", "1st - 4th grade",	"5th - 6th grade",	"7th - 8th grade",	"9th grade",	"10th grade",	"11th grade",	"High school graduate",	"Some college, No degree",	"Associate's degree, Occupational",	"Associate's degree, Academic",	"Bachelor's degree",	"Master's degree",	"Professional degree",	"Doctoral degree"]
black_ed_data.columns = rowlist
black_ed_data['None'] = black_ed_data['None'].fillna(0).astype(int)
black_ed_data
```

```python
black_ed_both = black_ed_data.iloc[6:20]
black_ed_male = black_ed_data.iloc[21:35]
black_ed_female = black_ed_data.iloc[36:50]
black_ed_both
```

```python
#Filter the data for the '25 years and over' row
target_row = black_ed_both.iloc[2]

#Prepare data for plotting
plot_data = target_row.drop(columns=['Total'])

#Transpose the data
plot_data_long = plot_data.T.reset_index()
plot_data_long.columns = ['Education Level', 'Count']
plot_data_long = plot_data_long.iloc[2:]

#Create the Lolli Chart
plt.figure(figsize=(10, 8))
sns.set_style("whitegrid")

#Create the plot
plt.hlines(y=plot_data_long['Education Level'],
           xmin=0, xmax=plot_data_long['Count'],
           color='skyblue', alpha=0.7, linewidth=2)

plt.scatter(plot_data_long['Count'], plot_data_long['Education Level'],
            color='dodgerblue', s=100, alpha=1)

#Final Touches
plt.title('Black Educational Attainment (All)', fontsize=15)
plt.xlabel('Count')
plt.tight_layout()
plt.show()
```
